```python
import numpy as np 
import pandas as pd 
from matplotlib.ticker import FuncFormatter

def vennootschapsbelasting(profit):
    schijf_1 = max(0, min(200000, profit))
    schijf_2 = max(0, profit-200000)
    return max(0, 0.20 * schijf_1 + 0.25 * schijf_2)


def vennootschapsbelasting_2021(profit):
    schijf_1 = max(0, min(200000, profit))
    schijf_2 = max(0, profit-200000)
    return max(0, 0.16 * schijf_1 + 0.20 * schijf_2)

df = pd.DataFrame(data={
    'vermogen': 1000000,
    'bruto rendement': 0,
    'belasting (2018)': 0,
    'effectief rendement (2018)': 0,
    'belasting (2021)': 0,
    'effectief rendement (2021)': 0
}, index=np.linspace(0, 20, 21))
df.index.name = 'rendement'

df['bruto rendement'] = df['vermogen'] * df.index / 100

for idx in df.index:
    profit = df.loc[idx, 'bruto rendement']
    df.loc[idx, 'belasting (2018)'] = int(vennootschapsbelasting(profit))
    df.loc[idx, 'belasting (2021)'] = int(vennootschapsbelasting_2021(profit))
    
df['effectief rendement (2018)'] = ( df['bruto rendement'] - df['belasting (2018)'] ) / df['vermogen'] * 100
df['effectief rendement (2021)'] = ( df['bruto rendement'] - df['belasting (2021)'] ) / df['vermogen'] * 100
```
